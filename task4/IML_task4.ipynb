```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('data/train_features.csv')
train_labels = pd.read_csv('data/train_labels.csv')
pretrain_data = pd.read_csv('data/pretrain_features.csv')
pretrain_labels = pd.read_csv('data/pretrain_labels.csv')
test_data = pd.read_csv('data/test_features.csv')
```

```python
train_data_feature = train_data.iloc[:, 2:]
pretrain_data_feature = pretrain_data.iloc[:, 2:]
test_data_feature = test_data.iloc[:, 2:]
train_labels_pd = train_labels.iloc[:,1:]
pretrain_labels_pd = pretrain_labels.iloc[:,1:]
```

```python
train_data_feature_np = train_data_feature.to_numpy()
pretrain_data_feature_np = pretrain_data_feature.to_numpy()
test_data_feature_np = test_data_feature.to_numpy()
train_labels_np = train_labels_pd.to_numpy()
pretrain_labels_np = pretrain_labels_pd.to_numpy()
```

```python
from sklearn.metrics import mean_squared_error

from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.layers import Add, Dense, Input, Dropout
from tensorflow.keras import optimizers, losses, activations, models, regularizers
import tensorflow as tf
from tensorflow.keras import backend as K
from tensorflow.keras.layers import Layer


from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
def linear():

    x_input = Input(shape=(1000))

    x_output = Dense(1000, activation='relu', kernel_initializer='normal',kernel_regularizer = regularizers.l1_l2(l1=1e-5, l2=1e-4))(x_input)

    #x_output = Dense(300, activation='relu', kernel_initializer='normal',kernel_regularizer = regularizers.l1_l2(l1=1e-5, l2=1e-4))(x_output)

    #x_output = Dense(100, activation='relu', kernel_initializer='normal',kernel_regularizer = regularizers.l1_l2(l1=1e-5, l2=1e-4))(x_output)
    x_output = Dense(1)(x_output)

    
    adam = tf.optimizers.Adam(learning_rate = 0.005)
    model = models.Model(inputs=x_input, outputs=x_output)
    

    model.compile(loss=tf.keras.losses.MeanSquaredError(), optimizer = adam, metrics=["RootMeanSquaredError"])
    model.summary()
    return model

Linear = linear()
```

```python
print(pretrain_data_feature_np.shape)
print(pretrain_labels_np.shape)
```

```python
earlystopping = EarlyStopping(monitor="val_loss", mode="min", patience=10, verbose=1)
redlrplat = ReduceLROnPlateau(monitor="val_loss", factor=0.5, mode="min", patience=5, verbose=2)
callbacks_list = [earlystopping, redlrplat] 
Linear.fit(pretrain_data_feature_np, pretrain_labels_np, 
               epochs = 1000, batch_size = 1000, verbose = 2, validation_split=0.1, callbacks=callbacks_list)
```

```python
# Freeze the previous layers
for layer in Linear.layers[:-1]:
    layer.trainable = False
```

```python
for layer in Linear.layers:
    print(layer, layer.trainable)
```

```python
inputs = Linear.inputs
x = (Linear.get_layer(index=1).output)
outputs = Dense(500, activation='relu',kernel_regularizer = regularizers.l1_l2(l1=1e-5, l2=1e-4))(x)
outputs = Dropout(rate=0.2)(outputs)
outputs = Dense(100, activation='relu',kernel_regularizer = regularizers.l1_l2(l1=1e-5, l2=1e-4))(outputs)
outputs = Dropout(rate=0.2)(outputs)
outputs = Dense(1)(outputs)
transferred_model = models.Model(inputs=inputs,outputs=outputs)
transferred_model.summary()
```

```python
for layer in transferred_model.layers:
    print(layer, layer.trainable)
```

```python
adam = tf.optimizers.Adam(learning_rate = 0.005)

transferred_model.compile(optimizer=adam, loss = tf.keras.losses.MeanSquaredError(),metrics=["RootMeanSquaredError"])
earlystopping = EarlyStopping(monitor="val_loss", mode="min", patience=50, verbose=1)
redlrplat = ReduceLROnPlateau(monitor="val_loss", factor=0.75, mode="min", patience=25, verbose=2)
callbacks_list = [earlystopping, redlrplat]
#, callbacks=callbacks_list
transferred_model.fit(train_data_feature_np, train_labels_np, 
               epochs = 500, batch_size = 8, verbose = 2, validation_split=0.2, callbacks=callbacks_list)
```

```python
test_data_feature_id_np = test_data.iloc[:, 0].to_numpy()
test_predict = transferred_model.predict(test_data_feature_np)
test_data_feature_id = pd.DataFrame(test_data_feature_id_np,columns = ['Id'])
test_predict_pd = pd.DataFrame(test_predict,columns = ['y'])
test_data_feature_id.reset_index(drop=True, inplace=True)
test_predict_pd.reset_index(drop=True, inplace=True)
prediction = pd.concat([test_data_feature_id,test_predict_pd], axis=1)
```
